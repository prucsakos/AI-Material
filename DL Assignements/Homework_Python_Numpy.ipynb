Az osztály neve legyen **Gradebook**. 6 publikus metódusa van:
* **add_student(student)**: Eltárolja a kapott diák nevét a naplóban, egy üres listát rendelve hozzá
* **get_all_students()**: Visszaadja a naplóban tárolt diákok neveit betűrendben
* **add_grade(student, grade)**: A jegyet beilleszti a kapott diákhoz tartozó listába. Ha nincs a naplóban ilyan nevű diák, *StudentMissingException*-t vált ki.
* **get_student_grades(student)**: Visszaadja a kapott diák jegyeit tartalmazó listát. Ha nincs a naplóban ilyan nevű diák, *StudentMissingException*-t vált ki.
* **get_avarage_grade_per_student()**: Visszaadja minden diákhoz a diák jegyeinek az átalgát.
* **get_students_with_many_5s()**: Visszaadja egy listában azon diákok neveit, akiknek több ötösük van, mint más jegyük.

### Megoldás

```
# Your solution -> 

class StudentMissingException(Exception):
  pass


class Gradebook:
  def __init__(self):
    self.grades = {}

  def add_student(self, student):
    if student not in self.grades.keys():
      self.grades[student] = []

  def get_all_students(self):
    result = [s for s in self.grades.keys()]
    result.sort()
    return result

  def add_grade(self, student, grade):
    if student in self.grades.keys():
      self.grades[student].append(grade)
    else:
      raise StudentMissingException
  
  def get_student_grades(self, student):
    if student in self.grades.keys():
      return self.grades[student]
    else:
      raise StudentMissingException

  def get_avarage_grade_per_student(self):
    result = {}
    summa = 0
    for student in self.grades.keys():
      
      summa = sum(self.grades[student])
      if summa  == 0:
        avg = 0
      else:
        avg = summa / len(self.grades[student])
      result[student] = avg
    return result

  def get_students_with_many_5s(self):
    result = []
    for student in self.grades.keys():
      counter = 0
      for grade in self.grades[student]:
        if grade == 5:
          counter += 1
      if counter > (len(self.grades[student])-counter):
        result.append(student)
    return result





```

## B/1

**Precision**

Multiclass klasszifikáció esetén a mintaelemeink címkéjét k kategória egyikébe becsüljük (például kutya, macska, papagáj, stb.). Ha a becsléshez neuronhálót használunk, tipikusan mintaelemenként egy-egy k elemű, valószínűségeket tartalmazó vektort kapunk, melyek a mintaelem egyes kategóriákba tartozásának valószínűségét reprezentálják. Ha szeretnénk ezekből a vektorokból megmondani, hogy egy-egy mintaelem melyik kategóriába tartozik a legnagyobb valószínűség szerint, akkor elég megmondani minden egyes vektorban a maximális elem indexét. Így megkapjuk a becsült kategóriákat.

A precision azt adja meg, hogy az adott kategóriába becsült mintaelemek milyen arányban tartoznak valóban ehhez a kategóriához. Azaz:

$$ Precision = \dfrac{VP}{VP+FP} $$

ahol VP a valós pozitívok (azaz ahol az adott kategóriát helyesen becsültük) száma, FP pedig a fals pozitívok száma (azaz ahol az adott kategóriát helytelenül becsültük). A nevező tehát egyenlő azzal, hogy az adott kategoriát hányszor becsültük.

## B/2 

**Recall**

Multiclass klasszifikáció esetén a mintaelemeink címkéjét k kategória egyikébe becsüljük (például kutya, macska, papagáj, stb.). Ha a becsléshez neuronhálót használunk, tipikusan mintaelemenként egy-egy k elemű, valószínűségeket tartalmazó vektort kapunk, melyek a mintaelem egyes kategóriákba tartozásának valószínűségét reprezentálják. Ha szeretnénk ezekből a vektorokból megmondani, hogy egy-egy mintaelem melyik kategóriába tartozik a legnagyobb valószínűség szerint, akkor elég megmondani minden egyes vektorban a maximális elem indexét. Így megkapjuk a becsült kategóriákat.

A recall azt adja meg, hogy az adott kategória egyedeit milyen arányban sikerült a megfelelő címkével ellátnunk. Azaz:

$$ Recall = \dfrac{VP}{VP+FN} $$

ahol VP a valós pozitívok száma (azaz ahol az adott kategóriát helyesen becsültük), FN pedig a fals negatívok száma (ahol nem az adott kategóriát becsültük, pedig az lett volna a helyes). A nevező tehát egyenlő azzal, hogy az adott kategória hányszor szerepel a helyes címkék közt.

Multiclass esetben minden kategóriára külön számoljuk, és ezeknek vesszük az átlagát.

A feladat, hogy implementáld a `recall` függvényt, ami két paramétert kap:
*   `y_pred` tartalmazza becsült valószínűségeket (ez egy (m, k) alakú tömb, ahol  m  a mintaelemek és k a kategóriák száma) 
*   `y_true` tartalmazza az igazi kategóriacímkéket (ez egy (m,) alakú tömb)

A függvény egy számot ad vissza, a kategóriákra a recall értékek átlagát.

**Kikötés:**  Az implementációt vektoros módon, NumPy-ban, ciklusok és egyéb, annak megfelelő Python konstrukciók használata nélkül kell elkészíteni. További részletek a notebook végén.

### Megoldás

```
import numpy as np

# Your solution ->
def recall( y_pred, y_true ):
  # VP + FN
  y_true_oh = np.zeros((y_true.size, y_true.max()+1))
  y_true_oh[np.arange(y_true.size),y_true] = 1
  vpfn = np.sum( y_true_oh, 0 )
  #Predictions one-hot
  y_pred_max = np.argmax( y_pred, 1 )
  y_pred_oh = np.zeros( (y_pred_max.size, y_pred_max.max()+1 ) )
  y_pred_oh[np.arange(y_pred_max.size), y_pred_max ] = 1

  #VP
  vp_oh = y_pred_oh * y_true_oh
  vp = np.sum( vp_oh, 0 )

  #result
  return np.mean( vp/vpfn )
```

Készíts egy **SudokuBoard** nevű osztályt, ami a Sudoku táblát reprezentálja! A Sudoku egy 9x9-es tábla, ami 3x3-as blokkokra van osztva a sorok és oszlopok mentén. 1-től 9-ig kell elhelyeznünk a számokat úgy, hogy minden sorban, oszlopban, és blokkban a számok egyszer szerepeljenek.

Az osztály a következő tagfüggvényekkel kell, hogy rendelkezzen:


*   `is_valid_placing(num, pos)`: `num` egy szám 1 és 9 között, `pos` pedig egy tuple amely két 0 és 8 közötti egész számot tartalmaz. A függvény logikai értéket ad vissza, amivel megadja, hogy az adott pozícióra elhelyezhető-e az adott szám.

*`get_square_with_most_numbers()`: visszaadja annak a négyzetblokknak az indexét, ahol a legtöbb szám van elhelyezve. A blokkokat 0-tól, sorfolytonosan indexeljük, tehát például a jobb felső sarokban lévő bokk indexe 2, a bal alsóé pedig 6.
*`get_possible_rows(num)`: visszaadja azon soroknak indexét (0-tól indexelve) egy tömbben, ahova az adott szám (`num` paraméter) elhelyezhető.
*`sum_of_missing_numbers_per_col()`: visszaadja oszloponként a hiányzó számok összegét egy tömbben.

**Kikötés:**  Az implementációt vektoros módon, NumPy-ban, ciklusok és egyéb, annak megfelelő Python konstrukciók használata nélkül kell elkészíteni. További részletek a notebook végén.




### Megoldás

```
import numpy as np

# Your solution ->
class SudokuBoard:
  def __init__(self):
    self.board = np.zeros((9,9))
  
  def is_valid_placing(self, num, pos):
    x,y = pos
    isempty = (self.board[x,y] == 0)
    hor = np.any(self.board[x,:] == num)
    ver = np.any(self.board[:,y] == num)
    bx = int(x/3)*3
    by = int(y/3)*3
    block = np.any(self.board[bx:bx+3, by:by+3] == num)
    return not (not isempty or hor or ver or block)


  def insert(self, num, pos):
    x,y = pos
    isfree = self.board[x,y] == 0
    if (isfree):
      self.board[x,y] = num
    return isfree

  def get_square_with_most_numbers(self):
    blocks = np.reshape(self.board, (3,3,3,3)).swapaxes(1,2).reshape(-1,3,3)
    mask = blocks > 0
    countList = np.sum(mask, axis=(1,2))
    max = np.argmax(countList, axis = 0)
    return max

  def get_possible_rows(self, num):
    rowList = self.board
    mask = rowList != num
    res = np.all(mask, axis = 1)
    return np.argwhere(res)
    
  def sum_of_missing_numbers_per_col(self):
    maxSum = 45
    columSum = np.sum(self.board, axis = 0)
    return maxSum - columSum
```

```
import unittest

class TestSudoku(unittest.TestCase):
    def test_is_valid_placing(self):
        sb = SudokuBoard()
        self.assertEqual(sb.board.shape, (9,9))
        self.assertTrue(np.all(sb.board == 0))
        self.assertTrue(sb.is_valid_placing(5, (0,0)))
        self.assertTrue(sb.is_valid_placing(5, (8,0)))
        self.assertTrue(sb.is_valid_placing(5, (0,8)))
        self.assertTrue(sb.is_valid_placing(5, (4,6)))
        self.assertTrue(sb.is_valid_placing(5, (8,8)))

        sb.board[3,3] = 5
        self.assertFalse(sb.is_valid_placing(7, (3,3)))
        self.assertFalse(sb.is_valid_placing(5, (5,4)))
        self.assertFalse(sb.is_valid_placing(5, (7,3)))
        self.assertFalse(sb.is_valid_placing(5, (3,7)))
        
        self.assertTrue(sb.is_valid_placing(5, (2,4)))
        self.assertTrue(sb.is_valid_placing(6, (3,4)))

    def test_insert(self):
        sb = SudokuBoard()
        self.assertTrue(sb.insert(4, (2,7)))
        self.assertEqual(sb.board[2,7], 4)
        self.assertFalse(sb.insert(4, (2,7)))

    def test_get_col_with_most_numbers(self):
        sb = SudokuBoard()
        sb.insert(3, (5,6))
        self.assertEqual(sb.get_col_with_most_numbers(), 6)
        
        to_insert = [(3, (1,7)),(6, (5,7)), (9, (2,1)),(1, (3,4))]
        [sb.insert(*t) for t in to_insert]
        self.assertEqual(sb.get_col_with_most_numbers(), 7)

        to_insert = [(4, (2,6)), (5, (6,6)), (2, (5,1))]
        [sb.insert(*t) for t in to_insert] 
        self.assertEqual(sb.get_col_with_most_numbers(), 6)

    def test_get_possible_squares(self):
        sb = SudokuBoard()
        possible_squares = sb.get_possible_squares(5)
        self.assertEqual(possible_squares.shape[1], 2)
        self.assertEqual(len(possible_squares), 9)
        
        sb.insert(5, (5,8))
        self.assertEqual(len(sb.get_possible_squares(5)), 8)

        sb.insert(6, (5,3))
        self.assertEqual(len(sb.get_possible_squares(5)), 8)
        self.assertEqual(len(sb.get_possible_squares(6)), 8)

        to_insert = [(5, (1,0)),(5, (2,7)), (5, (7,6))]
        [sb.insert(*t) for t in to_insert]
        self.assertEqual(len(sb.get_possible_squares(5)), 5)

    def test_sum_of_missing_numbers_per_row(self):
        sb = SudokuBoard()
        self.assertTrue(np.all(sb.sum_of_missing_numbers_per_row() == 45))

        to_insert = [(6, (6,5)),(7, (6,1)), (8, (6,3))]
        [sb.insert(*t) for t in to_insert]
        sum_missing = sb.sum_of_missing_numbers_per_row()
        self.assertEqual(sum_missing[6], 24)
        self.assertEqual(np.sum(sum_missing == 45), 8)

        to_insert = [(1, (3,0)),(2, (3,6)), (3, (3,8))]
        [sb.insert(*t) for t in to_insert]
        sum_missing = sb.sum_of_missing_numbers_per_row()
        self.assertEqual(sum_missing[3], 39)
        self.assertEqual(sum_missing[6], 24)
        self.assertEqual(np.sum(sum_missing == 45), 7)

def suite():
    suite = unittest.TestSuite()
    testfuns = ["test_is_valid_placing", "test_insert", 
                "test_get_col_with_most_numbers",
                "test_get_possible_squares",
                "test_sum_of_missing_numbers_per_row",
                ]
    [suite.addTest(TestSudoku(fun)) for fun in testfuns]
    return suite

runner = unittest.TextTestRunner(verbosity=2)
runner.run(suite())
```

# Kikötések

A hatékony, vektoros NumPy megvalósításhoz minimálisra kell csökkenteni a Python parancsok számát a kódban. Például, ha egy ciklusban hívunk NumPy függvényeket és a ciklus több milliószor lefut, az tipikusan jóval kevésbé hatékony, mint ha a Python ciklust megfelelő, vektorizált NumPy hívásokkal ki tudjuk váltani és a szkript lefutása során mindössze néhány, egyszerre nagyobb mennyiségű adaton operáló NumPy hívás fut le.

A ciklusok mellett az azt helyettesítő Python műveletek hatása is hasonló: a list/set/dict comprehension-ök, a `map(), filter()` függvények megvalósítása ugyancsak ciklussal történik. A NumPy-ban elérhető `np.apply_along_axis()` művelet egy adott n dimenziós tömbön operáló műveletet ismétel meg egy n+1 dimenziós tömb n dimenziós szeletein, majd összekonkatenálja az eredményt. Bár ezzel a művelettel kiváltható a Python ciklus használata, de sok esetben ez nem hatékony, így ilyenkor kerülendő ez a megközelítés. Például, ha az `np.apply_along_axis()` műveletnek megadott, ciklusban végrehajtott függvény Python változókat, vagy függvényeket használ, a megvalósítás nem lesz hatékonyabb egy egyszerű Python ciklusnál.

**Kikötések a 2. és 3. feladathoz:** 
* Nem használhatók Python ciklusok, comprehension-ök és generátorok (azaz tiltott a `for` és a `while` kulcsszavak használata).
* A feladatok nem oldhatók meg rekurzió segítségével.
* Nem használhatók a `map()` és `filter()` beápített függvények.
* Nem használhatók a következő, Python ciklust megvalósító, nem hatékony NumPy függvények: `np.apply_along_axis()`, `np.apply_over_axes()`, `np.vectorize()`, `np.frompyfunc()`.
* Bár a megoldásban nem kell más méretű táblákat kezelni, de a megoldás módjának skálázhatónak kell lennie a sorok, oszlopok száma (és a beléjük írható számjegyek száma) szerint. Azaz, nem elfogadható az a megközelítés, hogy ha kilenc fajta számjegyen szeretnénk végigiterálni, akkor a ciklus használatát úgy kerüljük el, hogy a kilenc számjegyet kilenc külön utasítással kezeljük, hiszen ekkor több sor/oszlop/számjegy esetén több sornyi kódra lenne szükségünk.
